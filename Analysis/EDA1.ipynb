```python
import pandas as pd

df = pd.read_csv('D:/EliteML-End-to-End-House-Price-Prediction-with-MLOps/Extracted_Data/AmesHousing.csv')
```

```python
numerical_features = df.select_dtypes(include=['int','float']).columns
```

```python
df_cleaned = df.copy()
df_cleaned[numerical_features] = df[numerical_features].fillna(df[numerical_features].mean())
```

```python
from sklearn.preprocessing import StandardScaler
df_cleaned[['Gr Liv Area','SalePrice']] = StandardScaler().fit_transform(df[['Gr Liv Area','SalePrice']])
```

```python
import numpy as np
threshold = 3
df_numeric = df_cleaned.select_dtypes(include=['int',"float"])
zscore = np.abs((df_numeric-df_numeric.mean())/df_numeric.std())
outliers = zscore> threshold
```

```python
df_cleaned = df_cleaned[(~outliers).all(axis=1)]
df_cleaned
```

```python
x = df_cleaned.drop(columns='SalePrice')
y = df_cleaned.SalePrice
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```
